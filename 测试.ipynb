```python
import torchaudio
import torch
from speechbrain.pretrained import SpeakerRecognition
import os
import pandas as pd
from tqdm.notebook import tqdm
```

```python
verification = SpeakerRecognition.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb", savedir="pretrained_models/spkrec-ecapa-voxceleb")
```

```python
src = "tests/samples/my_test/clean//"
files = os.listdir(src)
print(files)
```

```python
dict_={}
scores=[]
predictions = []
A = []
B = []
for i in tqdm(files):
    audio = verification.load_audio(os.path.join(src,i))
    embeddings = verification.encode_batch(audio)
    dict_[i] = embeddings
```

```python
dict_
```

```python
similarity = torch.nn.CosineSimilarity(dim=-1, eps=1e-6)
```

```python
for i in tqdm(files):
    # audio = verification.load_audio(os.path.join(src,i))
    # dict_[i] = audio
    for j in tqdm(files):
        if i != j:
            # audio1 = verification.load_audio(os.path.join(src,i))
            # audio2 = verification.load_audio(os.path.join(src,j))
            # score, prediction = verification.verify_batch(dict_[i], dict_[j],threshold=0.5) 
            score = similarity(dict_[i], dict_[j])
            A.append(i)
            B.append(j)
            scores.append(score)
            # predictions.append(prediction)
```

```python
scores_list = [float(i) for i in scores]
scores_list
```

```python
predictions_list = [str(bool(i)) for i in predictions]
predictions_list
```

```python
df = pd.DataFrame()
df["A"] = A
df["B"] = B
df["scores"] = scores_list
# df["predictions"] = predictions_list
df
```

```python
df.to_csv("results.csv")
```

```python
A_id = []
B_id = []

for i in df['A']:
    A_id.append(i.split("_")[0])
for i in df['B']:
    B_id.append(i.split("_")[0])
    
id_df = pd.DataFrame({"A-id":A_id,"B-id":B_id})
df = pd.concat([df,id_df],axis=1)
df
```

```python
A_id = df['A-id'].drop_duplicates()
B_id = df['B-id'].drop_duplicates()
```

```python
A_id
```

```python
df1 = pd.DataFrame()
for a_id in A_id:
    mean_dic = {}
    df2 = df[df['A-id']==a_id].sort_values(by=['scores'], ascending=False).head(3)
    df1 = pd.concat([df1,df2]).reset_index().drop(['index'],axis=1)
df1
```

```python
smi = []
for i in df1.T.items():
    a_id = str(i[1]['A-id']).split("_")[0]
    b_id = str(i[1]['B-id']).split("_")[0]
    if a_id!=b_id:
        smi_i=0
    else:
        smi_i=1
    smi.append(smi_i)
df1['smi']=smi
df1
```

```python
import numpy as np
result = []
for i in np.arange(0.4,0.99,0.1):
    len_p_all = len(df1['A-id'].drop_duplicates())-2
    rdf = df1[df1['scores']>=i]

    len_p = len(rdf[rdf['smi']==1]['smi'])
    result.append([round(i,2),len_p_all,len_p,str(round(len_p/len_p_all*100,2))+"%"])
xdf = pd.DataFrame(result,columns=["","","",""])
xdf
```
