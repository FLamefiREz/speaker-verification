```python
from Speaker.speakerlab.bin.infer import embedding
import math
import torch
import os
from tqdm.notebook import tqdm
import pandas as pd
import numpy as np
```

```python
embedding(wav_file="CMGAN/src/data/clean/1_1_1.wav").compute_embedding().shape
```

```python
src = "CMGAN/data/clean/"
files = os.listdir(src)
print(files)
```

```python
dict_={}
scores=[]
A = []
B = []
for i in tqdm(files):
    embeddings = embedding(wav_file=os.path.join(src,i)).compute_embedding()
    dict_[i] = embeddings
```

```python
dict_
```

```python
similarity = torch.nn.CosineSimilarity(dim=-1, eps=1e-6)
```

```python
for i in tqdm(files):
    for j in tqdm(files):
        if i != j:
            score = similarity(torch.tensor(dict_[i]), torch.tensor(dict_[j]))
            A.append(i)
            B.append(j)
            scores.append(score)
```

```python
scores_list = [float(i) for i in scores]
scores_list
```

```python
df = pd.DataFrame()
df["A"] = A
df["B"] = B
df["scores"] = scores_list
df
```

```python
A_id = []
B_id = []

for i in df['A']:
    A_id.append(i.split("_")[0])
for i in df['B']:
    B_id.append(i.split("_")[0])
    
id_df = pd.DataFrame({"A-id":A_id,"B-id":B_id})
df = pd.concat([df,id_df],axis=1)
df
```

```python
df.to_csv("results_all.csv")
```

```python
A_id = df['A-id'].drop_duplicates()
B_id = df['B-id'].drop_duplicates()
```

```python
A_id
```

```python
df1 = pd.DataFrame()
for a_id in A_id:
    mean_dic = {}
    df2 = df[df['A-id']==a_id].sort_values(by=['scores'], ascending=False).head(3)
    df1 = pd.concat([df1,df2]).reset_index().drop(['index'],axis=1)
df1
```

```python
smi = []
for i in df1.T.items():
    a_id = str(i[1]['A-id']).split("_")[0]
    b_id = str(i[1]['B-id']).split("_")[0]
    if a_id!=b_id:
        smi_i=0
    else:
        smi_i=1
    smi.append(smi_i)
df1['smi']=smi
df1
```

```python
df1.to_csv("results_top3.csv")
```

```python
result = []
for i in np.arange(0.4,0.99,0.1):
    len_p = 0
    len_p_all = len(A_id)
    rdf = df1[df1['scores']>=i]
    for j in A_id:
        if sum(rdf[rdf["A-id"]==j]["smi"])>=1:
            len_p+=1
    result.append([round(i,2),len_p_all,len_p,str(round(len_p/len_p_all*100,2))+"%"])
xdf = pd.DataFrame(result,columns=["","","",""])
xdf
```
